```python
import numpy as np
import pandas as pd
```

```python
training_set_features = r"/Users/aayushjain/codes/project/IIT Indore/training_set_features.csv"
X = pd.read_csv(training_set_features,index_col="respondent_id")

#Loading the target label dataset
training_set_labels = r"/Users/aayushjain/codes/project/IIT Indore/training_set_labels.csv"
y = pd.read_csv(training_set_labels,index_col="respondent_id")
```

```python
print("features_df.shape", X.shape)
X.head()
```

```python
X.dtypes
```

```python
X_new = X.fillna(-999)
pd.isnull(X_new).any()
```

```python
test_set_features = r"/Users/aayushjain/codes/project/IIT Indore/test_set_features.csv"
test_file = pd.read_csv(test_set_features,index_col="respondent_id")
test_file.head()
```

```python
test_file_new = test_file.fillna(-999)
pd.isnull(test_file_new).any()
```

```python
y_h1n1=y['h1n1_vaccine']
y_seas=y['seasonal_vaccine']
```

```python
from catboost import CatBoostClassifier
cat_feat=['age_group','education','race','sex', 'income_poverty', 'marital_status','rent_or_own','employment_status','hhs_geo_region','census_msa','employment_industry','employment_occupation']
model=CatBoostClassifier(loss_function='Logloss',cat_features=cat_feat)

model.fit(X_new,y_h1n1) 
pp1=model.predict_proba(test_file_new)

model.fit(X_new,y_seas) 
pp2=model.predict_proba(test_file_new)
```

```python
from sklearn.metrics import accuracy_score
y_pred_h1n1 = model.predict(X_new)
y_pred_h1n1_binary = np.round(y_pred_h1n1)
accuracy_h1n1 = accuracy_score(y_h1n1, y_pred_h1n1_binary)
y_pred_seas = model.predict(X_new)
y_pred_seas_binary = np.round(y_pred_seas)
accuracy_seas = accuracy_score(y_seas, y_pred_seas_binary)
print(f"H1N1 Vaccine Accuracy: {accuracy_h1n1 * 100:.2f}%")
print(f"Seasonal Vaccine Accuracy: {accuracy_seas * 100:.2f}%")
```

```python
# Check data types of features in X_new
print(X_new.dtypes)

# Check for NaN values
print(X_new.isnull().sum())

# Check the shape
print(X_new.shape)
# Set categorical data type for relevant columns
categorical_columns = ['age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa', 'employment_industry', 'employment_occupation']

X_new[categorical_columns] = X_new[categorical_columns].astype('category')
```

```python

# Assuming model is your trained Keras Sequential model
y_pred_probabilities = model.predict(X_new)

# Assuming you want the probabilities for class 1 (positive class)
# If you need probabilities for a different class, adjust the index accordingly
pp1 = y_pred_probabilities[:, 1]

```

```python
print(pp2)
```

```python
sub_for = r"/Users/aayushjain/codes/project/IIT Indore/submission_format.csv"
submission_df = pd.read_csv(sub_for,index_col="respondent_id")
m=[]
for i in pp1:
  m.append(i[1])
print(m)

n=[]
for i in pp2:
  n.append(i[1])
print(n)
submission_df["h1n1_vaccine"] = m
submission_df["seasonal_vaccine"] = n
submission_df.to_csv(r'submission_final.csv', index=True)
predictions_h1n1 = []

```

```python
for i in range(10):
    clf = CatBoostClassifier(learning_rate=0.1, iterations=1000, random_seed=i, logging_level='Silent')
    clf.fit(X_new, y_h1n1, cat_features=cat_feat)
    predictions_h1n1.append(clf.predict_proba(test_file_new)[:,1])


```

```python
predictions_seas = []

for i in range(10):
    clf = CatBoostClassifier(learning_rate=0.1, iterations=1000, random_seed=i, logging_level='Silent')
    clf.fit(X_new, y_seas, cat_features=cat_feat)
    predictions_seas.append(clf.predict_proba(test_file_new)[:,1])
prediction_h1n1 = np.mean(predictions_h1n1, axis=0)
prediction_seas = np.mean(predictions_seas, axis=0)
pd.DataFrame(
    {'h1n1_vaccine':prediction_h1n1,'seasonal_vaccine':prediction_seas}
).to_csv(
    'submission_improved_bagged.csv', index=True
)
```
